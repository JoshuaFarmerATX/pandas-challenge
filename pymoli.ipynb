## Age Demographics
* Establish bins for ages
* Categorize the existing players using the age bins. Hint: use pd.cut()
* Calculate the numbers and percentages by age group
* Create a summary data frame to hold the results
* Optional: round the percentage column to two decimal points
* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
bin_names = ["<10", "10 - 14", "15 - 19", "20 - 24", "25 - 29", "30 - 34", "35 - 39", "40+"]
df["Age Group"] = pd.cut(df['Age'], bins, labels = bin_names)
age_demo_df = df.drop_duplicates(subset = "SN")
```

```python
age_analysis = {
    "Total Count": [len(age_demo_df[["SN", "Age"]].loc[age_demo_df["Age Group"] == bin_names[0]]), len(age_demo_df[["SN", "Age"]].loc[age_demo_df["Age Group"] == bin_names[1]]), len(age_demo_df[["SN", "Age"]].loc[age_demo_df["Age Group"] == bin_names[2]]), len(age_demo_df[["SN", "Age"]].loc[age_demo_df["Age Group"] == bin_names[3]]), len(age_demo_df[["SN", "Age"]].loc[age_demo_df["Age Group"] == bin_names[4]]), len(age_demo_df[["SN", "Age"]].loc[age_demo_df["Age Group"] == bin_names[5]]), len(age_demo_df[["SN", "Age"]].loc[age_demo_df["Age Group"] == bin_names[6]]), len(age_demo_df[["SN", "Age"]].loc[age_demo_df["Age Group"] == bin_names[7]])],
    "Percentage of Players": [str(round((len(age_demo_df[["SN", "Age"]].loc[age_demo_df["Age Group"] == bin_names[0]]) / len(age_demo_df[["SN"]]) * 100), 2)) + "%", str(round((len(age_demo_df[["SN", "Age"]].loc[age_demo_df["Age Group"] == bin_names[1]]) / len(age_demo_df[["SN"]]) * 100), 2)) + "%", str(round((len(age_demo_df[["SN", "Age"]].loc[age_demo_df["Age Group"] == bin_names[2]]) / len(age_demo_df[["SN"]]) * 100), 2)) + "%", str(round((len(age_demo_df[["SN", "Age"]].loc[age_demo_df["Age Group"] == bin_names[3]]) / len(age_demo_df[["SN"]]) * 100), 2)) + "%", str(round((len(age_demo_df[["SN", "Age"]].loc[age_demo_df["Age Group"] == bin_names[4]]) / len(age_demo_df[["SN"]]) * 100), 2)) + "%", str(round((len(age_demo_df[["SN", "Age"]].loc[age_demo_df["Age Group"] == bin_names[5]]) / len(age_demo_df[["SN"]]) * 100), 2)) + "%", str(round((len(age_demo_df[["SN", "Age"]].loc[age_demo_df["Age Group"] == bin_names[6]]) / len(age_demo_df[["SN"]]) * 100), 2)) + "%", str(round((len(age_demo_df[["SN", "Age"]].loc[age_demo_df["Age Group"] == bin_names[7]]) / len(age_demo_df[["SN"]]) * 100), 2)) + "%"]
}
age_df = pd.DataFrame(age_analysis)
age_df.index = bin_names
age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age
* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
def column_total():
    column_output = []
    for row in range(len(df["Age Group"].unique())):
        column_output.append(len(df["SN"].loc[df["Age Group"] == bin_names[row]]))
    return column_output

def total_purchase_price():
    column_output = []
    for row in range(len(df["Age Group"].unique())):
        column_output.append(usd_format(df["Price"].loc[df["Age Group"] == bin_names[row]].sum()))
    return column_output

def ave_purchase_price():
    column_output = []
    for row in range(len(df["Age Group"].unique())):    
        column_output.append(usd_format((df["Price"].loc[df["Age Group"] == bin_names[row]].sum()) / (len(df["SN"].loc[df["Age Group"] == bin_names[row]]))))
    return column_output

def ave_per_person():
    column_output = []
    for row in range(len(df["Age Group"].unique())):
        column_output.append(usd_format((df["Price"].loc[df["Age Group"] == bin_names[row]].sum()) / (len(age_demo_df["SN"].loc[age_demo_df["Age Group"] == bin_names[row]]))))
    return column_output

purchasing_analysis_age = {
    "Purchase Count": column_total(),
    "Average Purchase Price": ave_purchase_price(),
    "Total Purchase Price": total_purchase_price(),
    "Average Total Purchase Per Person": ave_per_person()
}

purchasing_analysis_df = pd.DataFrame(purchasing_analysis_age)
purchasing_analysis_df.index = bin_names
purchasing_analysis_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below
* Create a summary data frame to hold the results
* Sort the total purchase value column in descending order
* Optional: give the displayed data cleaner formatting
* Display a preview of the summary data frame

```python
def ts_purch_count():
    column_output = []
    for row in df["SN"].unique():
        column_output.append(len(df["SN"].loc[df["SN"] == row]))
    return column_output

def ts_ave_purch():
    column_output = []
    for row in df["SN"].unique():
        column_output.append(usd_format((df["Price"].loc[df["SN"] == row].sum()) / (len(df["SN"].loc[df["SN"] == row]))))
    return column_output

def ts_total_purch_price():
    column_output = []
    for row in df["SN"].unique():
        column_output.append(usd_format(df["Price"].loc[df["SN"] == row].sum()))
    return column_output

def ts_total_purch_price_sort():
    column_output = []
    for row in df["SN"].unique():
        column_output.append(df["Price"].loc[df["SN"] == row].sum())
    return column_output

top_spenders = {
    "SN": df["SN"].unique(),
    "Purchase Count": ts_purch_count(),
    "Average Purchase Price": ts_ave_purch(),
    "Total Purchase Price": ts_total_purch_price(),
    "Total Purchase Price Sort": ts_total_purch_price_sort()
}
top_spenders_df = pd.DataFrame(top_spenders).set_index("SN").sort_values("Total Purchase Price Sort", ascending = False)
top_spenders_df = top_spenders_df.drop(columns = "Total Purchase Price Sort")
top_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns
* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
* Create a summary data frame to hold the results
* Sort the purchase count column in descending order
* Optional: give the displayed data cleaner formatting
* Display a preview of the summary data frame

```python
mpi = df[["Item ID", "Item Name", "Price"]]
mpi_grouped = mpi.groupby(["Item ID", "Item Name"]).mean()
mpi_count = mpi.groupby(["Item ID", "Item Name"]).count()
mpi_grouped.insert(0, "Purchase Count", mpi_count)
mpi_grouped.insert(2, "Total Purchase Value", mpi.groupby(["Item ID", "Item Name"]).sum())
mpi_grouped['Price'] = mpi_grouped['Price'].map('${:,.2f}'.format)
mpi_grouped['Total Purchase Value'] = mpi_grouped['Total Purchase Value'].map('${:,.2f}'.format)
mpi_grouped.sort_values("Purchase Count", ascending = False).head()
```
